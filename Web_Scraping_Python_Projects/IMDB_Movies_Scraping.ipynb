```python
import requests,pandas as pd
from bs4 import BeautifulSoup
url1 = "https://www.imdb.com/list/ls091520106/"
headers = {
    "authority": "www.imdb.com","method": "GET","path": "list/ls091520106/","scheme": "https",
    "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "accept-encoding": "gzip, deflate, br","accept-language": "en-US,en;q=0.9","cache-control": "max-age=0",
    "sec-fetch-dest": "document","sec-fetch-mode": "navigate","sec-fetch-site": "none","sec-fetch-user": "?1","upgrade-insecure-requests": "1",
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url1, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
print("Request Granted !!!!",response.status_code) if response.status_code == 200 else print("Request Denied !!!!",response.status_code)
top_100_movies = soup.find_all("h3", {"class": "ipc-title__text"}) if soup.find_all("h3", {"class": "ipc-title__text"}) else "N/A"
release_year = soup.find_all(
    "span", {"class": "sc-b189961a-8 hCbzGp dli-title-metadata-item"}) if soup.find_all(
    "span", {"class": "sc-b189961a-8 hCbzGp dli-title-metadata-item"}) else "N/A"
ratings = soup.find_all("span", {"class": "ipc-rating-star--rating"}) if soup.find_all("span", {"class": "ipc-rating-star--rating"}) else "N/A"
movielist = []
for title in top_100_movies: movielist.append(title.get_text(strip = True))#print(title.get_text(strip = True))
yearlist = []
for year in release_year: yearlist.append(year.get_text(strip = True))
ratinglist = []
for r in ratings: ratinglist.append(r.get_text(strip= True))
#print("Movie Titles : \n",movielist[0:25])
#print("Years : \n",yearlist[0::3])
#print("Ratings : \n",ratinglist)
#print("Certifications : \n",yearlist[2::3])
#print(len(movielist[0:25]),len(yearlist[0::3]),len(ratinglist),len(yearlist[2::3]))
#print(year.get_text(strip = True))
rank=[]
names2=[]
nm=[]
for i in movielist:nm.append(i.split("."))
for i in range(0,len(nm)):rank.append(nm[i][0])
for i in range(0,len(nm)-2):names2.append(nm[i][1])
movies = pd.DataFrame({})
movies['Rank'] = rank[0:25]
movies['Title'] = names2
movies['Release Year'] = yearlist[0::3]
movies['IMDB Rating'] = ratinglist
movies['Certification'] = yearlist[2::3]
movies_csv = movies.to_csv('list_of_movies.csv',index=False)
print("Data has been written to list_of_movies.csv")
```
