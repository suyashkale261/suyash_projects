```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 100)
y = x * 2
z = x ** 2
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot(x, y, label='y = x * 2')
ax1.set_title('Plot of x vs y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax2.plot(x, z, label='z = x^2', color='r')
ax2.set_title('Plot of x vs z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.legend()
plt.tight_layout()
plt.show()
```
