```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 100)
y = x * 2
z = x ** 2
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot(x, y, label='y = x * 2')
ax1.set_title('Plot of x vs y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax2.plot(x, z, label='z = x^2', color='r')
ax2.set_title('Plot of x vs z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
titanic_data = pd.read_csv("titanic_original.csv")
gender_counts = titanic_data['sex'].value_counts()

# (i) Create a pie chart presenting the male/female proportion
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'pink'])
plt.title('Male/Female Proportion on the Titanic')
plt.show()

# (ii) Create a scatterplot with the Fare paid and the Age, differentiated by gender
plt.figure(figsize=(10, 6))
colors = {'male': 'blue', 'female': 'red'}
plt.scatter(titanic_data['age'], titanic_data['fare'], c=titanic_data['sex'].map(colors), alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare paid vs Age, differentiated by Gender')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Male'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Female')])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("diamonds.csv")
filtered_df = df[(df['clarity'] == 'SI2') & (df['color'] == 'E')]
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=filtered_df, x='carat', y='price', hue='cut', palette='viridis', s=100, edgecolor='w', alpha=0.75)
plt.title('Relationship Between Carat and Price for Diamonds with Clarity "SI2" and Color "E"', fontsize=18)
plt.xlabel('Carat (Size of Diamond)', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.legend(title='Cut', title_fontsize='13', fontsize='11', loc='upper left')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
planets_df = pd.read_csv("planets.csv")
masses = planets_df['mass']

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
# Histogram
ax[0].hist(masses, bins=30, color='skyblue', edgecolor='black')
ax[0].set_title('Distribution of Planet Masses', fontsize=15)
ax[0].set_xlabel('Mass (Jupiter Masses)', fontsize=12)
ax[0].set_ylabel('Frequency', fontsize=12)
#Box Plot
ax[1].boxplot(masses, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'))
ax[1].set_title('Box Plot of Planet Masses', fontsize=15)
ax[1].set_ylabel('Mass (Jupiter Masses)', fontsize=12)

fig.suptitle('Extrasolar Planets Mass Distribution', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
