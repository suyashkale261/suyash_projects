```python
result = soup.find_all('span',attrs={"class":"short-desc"})
```

```python
# find() -> searches for the first matching tag and returns the tag object
# find_all() -> searches for all the matches and return the resulted list

# attributes -> text and content

firstres = result[0]
firstres.find('strong')
```

```python
record = []
for i in result:
    date = i.find('strong').text[:-1]+',2017'
    lie = i.contents[1][1:-2]
    exp = i.contents[2].find('a').text[1:-1]
    url = i.contents[2].find('a')['href']
    record.append((date,lie,exp,url))
record
```

```python
import pandas as pd
col_name = ["Date","Lie","Explaination","URL"]
df = pd.DataFrame(record,columns = col_name)
df
```

```python
url = 'https://www.imdb.com/list/ls091520106/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36'
}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    # Find all h3 tags with the specified class
    h3_tags = soup.find_all('h3', class_='ipc-title__text')
    for h3 in h3_tags:
        # Get the text content inside the h3 tag
        print(h3.get_text(strip=True))
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
names = soup.find_all('h3',class_="ipc-title__text")
names
```

```python
nameslist = []
for i in names:nameslist.append(i.text.replace("\n",""))
nameslist
```
