```python
%matplotlib inline
mean = data[data["normalized-losses"]!="?"]["normalized-losses"].astype(dtype = "float").mean()
plt.hist(data[data["normalized-losses"]!="?"]["normalized-losses"].astype(dtype = "float"),bins=10)
```

```python
mean = data[data["price"]!="?"]["price"].astype(dtype = "float").mean()
plt.hist(data[data["price"]!="?"]["price"].astype(dtype = "float"),bins=10)
```

```python
data['normalized-losses'] = data['normalized-losses'].str.replace("?",str(mean)).astype(dtype="float")
```

# WHEN AND WHERE TO USE MEAN AND MEDIAN ALWAYS DEPENDS ON YOUR DATASET AND TASK
# IF THE DATASET HAS TOO MANY OUTLIERS, WE PREFER MEDIAN
# EG. 99% OF HOUSEHOLD INCOME IS BELOW 100 AND 1% IS ABOVE 500

```python
median = data[data['price']!="?"]['price'].astype(dtype = "float").median()
data['price']=data['price'].replace("?",(median)).astype(dtype='float')
data['horsepower'].value_counts()
```

```python
median_hp = data[data["horsepower"]!="?"]["horsepower"].astype(dtype = "float").median()
data['horsepower']=data['horsepower'].replace("?",str(median_hp)).astype(dtype='float')
data['horsepower'].value_counts()
```

```python
mean_rpm = data[data['peak-rpm']!='?']['peak-rpm'].astype(dtype='float').mean()
data['peak-rpm'] = data['peak-rpm'].str.replace('?',str(mean_rpm)).astype(dtype="float")
data['peak-rpm'].value_counts()
```

```python
mean_bore = data[data['bore']!='?']['bore'].astype(dtype='float').mean()
data['bore']=data['bore'].str.replace('?',str(mean_bore)).astype(dtype='float')
data['bore'].value_counts()
```

```python
mode_n= data[data['num-of-doors']!='?']['num-of-doors'].mode()
data['num-of-doors'] = data['num-of-doors'].str.replace('?',str(mode_n))
data['num-of-doors'].value_counts()
```

```python
mode=data[data["engine-type"]!= "|"]["engine-type"].mode()
data['engine-type'] = data['engine-type'].str.replace('?',str(mode_n))
data['engine-type'].value_counts()
```
